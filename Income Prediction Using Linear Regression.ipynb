```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#STEP 2: LOAD THE DOCUMENTS
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
print('First 5 rows:')
print(df.head())
```

```python
df
```

```python
#STEP 3: Prepare the Data (x and y)
```

```python
X = df[['year']]  # Feature
y = df['per capita income (US$)']  # Target
```

```python
#STEP 4: Train and create model
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
# ðŸ“Œ Step 5: User Input for Year Prediction (Improved)
try:
    custom_year = int(input("ðŸ“… Enter the year you want to predict income for (e.g., 2025): "))
    predicted_income = model.predict(pd.DataFrame({'year': [custom_year]}))
    print(f"\nðŸ’° Predicted income in {custom_year}: ${predicted_income[0]:,.2f}")
except ValueError:
    print("âš ï¸ Please enter a valid integer year.")


# ðŸ“Œ Step 6: Visualize the Data and the Model
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, model.predict(X), color='red', label='Regression Line')
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.title('Canada Per Capita Income Prediction (Linear Regression)')
plt.legend()
plt.grid(True)
plt.show()
```
